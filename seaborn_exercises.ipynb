```python

# Using the chipotle dataset from the previous exercise, 
# create a bar chart that shows the 4 most popular items and the revenue produced by each.
```

```python
def get_db_url(db_name):
    from env import host, user, password
    return f'mysql+pymysql://{user}:{password}@{host}/{db_name}'
    
chipotle = get_db_url('chipotle')
```

```python
sql = """
    SELECT * FROM orders
    """

Chipotle_DF = pd.read_sql(sql, chipotle)
```

```python
Chipotle_DF['price'] = Chipotle_DF['item_price'].str.replace('$','').astype(float)
```

```python
Chipotle_DF.info()
```

```python
Chipotle_DF.head()
```

```python
Top_Orders = Chipotle_DF.groupby(['item_name']).quantity.agg('sum')
coolgraph =Top_Orders.nlargest(4)
coolgraph.index
```

```python
coolgraph.values
```

```python
sns.barplot(x = coolgraph.values , y = coolgraph.index)
```
