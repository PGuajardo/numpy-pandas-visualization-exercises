```python
import matplotlib.pyplot as plt  

#import other libraries 
import numpy as np
import math
from random import randint
import random
import pandas as pd
```

```python
n = range(1,6)

#y1 = n*0 + 1
y1 = [i*0 + 1 for i in n]

#y2 = log(n)
y2= [math.log(i) for i in n]

#y3 = n
y3 = [i for i in n]

#y4 = n* y2
y4 = [i * math.log(i) for i in n]

#y5 = n^2
y5 = [i**2 for i in n]

#y6 = 2^n
y6 = [2**i for i in n]

#y7 = n!
fact = 1
y7 = [fact * i for i in n]

plt.figure(figsize = (10,10))
plt.grid('True', ls = '--')

plt.title("Big O Notation")
plt.xlabel("Elements")
plt.ylabel("Operations")

plt.plot(n,y1, label = '$O(1)$')

plt.plot(n,y2, label = '$O(log n)$')

plt.plot(n,y3, label = '$O(n)$')

plt.plot(n,y4, label = '$O(n log n)$')

plt.plot(n,y5, label = '$O(n^2)$')

plt.plot(n,y6, label = '$O(2^n)$')

plt.plot(n,y7, label = '$O(n!)$')

plt.ylim(0,16)
plt.xlim(1,5)
plt.legend()
```
