# 1.)Use matplotlib to plot the following equation:
`y=x2−x+2`
You'll need to write the code that generates the x and y points.
Add an anotation for the point 0, 0, the origin.



```python
import matplotlib.pyplot as plt  

#import other libraries 
import numpy as np
import math
from random import randint
import random
import pandas as pd
```

```python
x = range(-100,100)
y = [((n**2) - (n+2)) for n in x]


plt.figure(figsize = (10,8))
plt.title('$ y = x^2 - x +2$')

plt.xlabel('$x$')
plt.ylabel('$y$')

plt.plot(x,y)
plt.annotate('Origin',
             xy = (0,0),
             xytext = (5,2000),
             arrowprops = dict(facecolor = 'grey')
            )
```

# 4.) Combine the figures you created in the last step into one large figure

```python
x1 = range(10)
y1 = [math.sqrt(x) for x in x1]
#y=√x

y2 = [x**3 for x in x1]
#y=x3

y3 = [2**x for x in x1]
#y=2x

y4 = [float(1/(x+1)) for x in x1]
#y=1/(x+1)

plt.figure(figsize = (10,10))
plt.ylim(0, 10)
plt.xlim(0, 10)

plt.plot(x1,y1, c = 'red', label = '$Sqrt(x)$')
plt.plot(x1,y2, c = 'blue', label = '$x^3$')
plt.plot(x1,y3, c = 'green', label = '$2 ^ x$')
plt.plot(x1,y4, c = 'orange', label = '$Float x$')


plt.ylabel('$y$')
plt.xlabel('$x$')
plt.legend()
plt.title("Variations of X")
```
