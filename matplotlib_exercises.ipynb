```python

x1 = range(1,20)
y1 = [math.sqrt(x) for x in x1]
#y=âˆšx

y2 = [x**3 for x in x1]
#y=x3

y3 = [2**x for x in x1]
#y=2x

y4 = [float(1/(x+1)) for x in x1]
#y=1/(x+1)

plt.figure(figsize = (10,10))

plt.subplot(2,2,1)
plt.plot(x1,y1)
plt.xlabel("x axis")
plt.ylabel("y axis")


plt.subplot(2,2,2)
plt.plot(x1,y2)
plt.xlabel("x axis")
plt.ylabel("y axis")


plt.subplot(2,2,3)
plt.plot(x1,y3)
plt.xlabel("x axis")
plt.ylabel("y axis")

plt.subplot(2,2,4)
plt.plot(x1,y4)
plt.xlabel("x axis")
plt.ylabel("y axis")
```

# 4.) Combine the figures you created in the last step into one large figure

```python
x1 = range(5)
y1 = [math.sqrt(x) for x in x1]
#y=âˆšx

y2 = [x**3 for x in x1]
#y=x3

y3 = [2**x for x in x1]
#y=2x

y4 = [float(1/(x+1)) for x in x1]
#y=1/(x+1)

plt.figure(figsize = (10,10))


plt.plot(x1,y1, c = 'red', label = '$Sqrt(x)$')
plt.plot(x1,y2, c = 'blue', label = '$x^3$')
plt.plot(x1,y3, c = 'green', label = '$2 ^ x$')
plt.plot(x1,y4, c = 'orange', label = '$Float x$')

plt.legend()
plt.title("Variations of X")
```
